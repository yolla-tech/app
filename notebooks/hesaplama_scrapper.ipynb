```python
class ScrapperPayload:
    def set_data(self, data: ScrapperData) -> None:
        self.data = data
    
    def set_payload(
        self,
        city_a: str, 
        city_b: str,
        weight: float = None, 
        width: float = None, 
        height: float = None, 
        length: float = None,
        type: str = "letter", 
        from_address: bool = False, 
        to_address: bool = False) -> None:
        
        if type != "letter" and type != "box":
            raise Exception(f"Type must be either letter or box: {type}")
    
        if type == "box" and (weight is None or width is None or length is None):
            raise Exception(f"Box must have weight, width, and height: Type->{type} Width->{width} Height->{height} Length->{length}")

        self.payload = {
            "sehira": self.data.cities[city_a],
            "sehirb": self.data.cities[city_b],
            "agirlik": weight,
            "en": width,
            "boy": length,
            "yukseklik": height,
        }
    
        if from_address:
            self.payload["adrestenalim"] = "E"
        
        if to_address:
            self.payload["adreseteslim"] = "E"
            
        if type == "letter":
            self.payload["gonderituru"] = "D"
```

```python
def scrap_data(payload: ScrapperPayload, data: ScrapperData):
    def format_html(contents: list):
        results = []

        i = 0
        while i < len(contents):
            element = contents[i]
            
            # Check if the element is a <b> tag (service name)
            if isinstance(element, Tag) and element.name == 'b':
                service_name = element.get_text(strip=True)
                
                # Check if this is the 'NOT' section
                if service_name == 'NOT:':
                    # Extract the note text
                    note_text = contents[i + 1].strip()
                    break  # Exit the loop after handling the note
                else:
                    # Get the price info from the next element
                    price_info = contents[i + 1].strip()
                    # Append the service name and price to the results list
                    results.append({'service': service_name, 'price': price_info})
                    i += 2  # Skip the <br/> tag
            i += 1  # Move to the next element
        
        services = {}

        for item in results:
            name = item['service'].split()[0]
            price = float(item['price'].split()[0].replace(',', '.'))
            services[name] = min(services.get(name, math.inf), price)
        
        return services

    response = data.session.post(data.url, headers=data.headers, data=payload.payload, cookies=data.cookies)
    if response.status_code != 200:
        raise Exception(f"Failure, status code: {response.status_code}")
    soup2 = BeautifulSoup(response.content, "html.parser")
    return format_html(soup2.find("div", {"id": "hsonuc"}).contents)
```

```python
payload = ScrapperPayload()
data = ScrapperData()
payload.set_data(data)
payload.set_payload(city_a="adana", city_b="ankara", weight=20, width=20, height=20, length=20, type="box", from_address=True, to_address=True)
scrap_data(payload, data)
```
